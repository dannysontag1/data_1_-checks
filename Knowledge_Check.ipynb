```python
df = pd.DataFrame(json_arr)
df
```

```python
# pull out current temp from the dictionary and put this value in the field
df['main'] = df['main'].str['temp']

# Convert from kelvin to fahrenheit
df['main'] = (df['main'] - 273.15) * (9/5) + 32

# pull out the country and put it back in the field
df['sys'] = df['sys'].str['country'].astype(str)
df
```

```python
# Pull out the columns that I want and put in logical Order
df = df[['name', 'sys', 'main']]
df
```

```python
# Renaming Columns
df.columns = ['City', 'Country', 'Current Temp (F)']
df
```

```python
#Sort Values From High Temp to Low
df.sort_values('Current Temp (F)', ascending=False)
```

```python
# Only returns Cities where temperature over 60 fahrenheit

df.loc[df['Current Temp (F)'] > 60].sort_values('Current Temp (F)', ascending=False)
```
